```

# Feistel Cipher 
PT = input()
 
  
# Converting the plain text to 
# ASCII 
PT_Ascii = [ord(x) for x in PT] 
  
# Converting the ASCII to  
# 8-bit binary format 
PT_Bin = [format(y,'08b') for y in PT_Ascii] 
PT_Bin = "".join(PT_Bin) 
  
n = int(len(PT_Bin)//2) 
L1 = PT_Bin[0:n] #left half 
R1 = PT_Bin[n::] #right half
m = len(R1) 
   
# Generate Key K1 for the  
# first round 
K1= rand_key(m)  #length of key should be same as R1
   
# Generate Key K2 for the 
# second round 
K2= rand_key(m) 
   
# first round of Feistel 
f1 = exor(R1,K1) 
R2 = exor(f1,L1) 
L2 = R1 
   
# Second round of Feistel 
f2 = exor(R2,K2) 
R3 = exor(f2,L2) 
L3 = R2 
   
# Cipher text 
bin_data = L3 + R3 
str_data =' '
print(len(bin_data))
print(bin_data)
  
```
