```
#function to genearte inital subkey (wo,w1,w2,w3)
def generate_subkey0(key_l):
  word_matrix=np.array(key_l)
  word_matrix=word_matrix.reshape(4,4)
  df=pd.DataFrame(word_matrix.T,columns=['w0','w1','w2','w3'])
  return word_matrix,df
```

```
#hex-word in hex
def rotword(word_matrix,i):
  word=word_matrix[i-1,:]
  #print(word,i)
  word=list(word)
  rotword=word[1:]+word[:1]
  return rotword
```

```
def subword(rotword):
  sub_list=[]
  sub_list_hex_bit=[]
  for i in rotword:
    for j in range(0,len(i)-1):
      sub_list.append(sbox[int(i[j],16)][int(i[j+1],16)]) #sub_list in int
  #sub_list_bin=["{0:08b}".format(int(hex(t), 16)) for t in sub_list]
  for i in sub_list:
      value=hex(i).replace('0x','')
      if len(value)==1:
        value='0'+value
      sub_list_hex_bit.append(value)

  #sub_list_hex_bit=['0'+hex(t).replace('0x','') if len(sub_list)<8 else hex(t).replace('0x','') for t in sub_list ]
  #sub_list_hex_bytes=groupbytes(sub_list_hex_bit)
  return ''.join(sub_list_hex_bit)
```

```
def groupbytes(key):
  key_l=[]
  try:
    for i in range(0,len(key),2):
      key_l.append(key[i]+key[i+1])
      #print(key_l)
  except:
    print('')
  return key_l
```

```
def xor(subword,rcon):
  #lb_sub=''.join(format( i,'08b') for i in subword)int to binary
  subword_hextobin="{0:08b}".format(int(subword, 16))
  rcon_hextobin="{0:08b}".format(int(rcon, 16)) #hex to binary
  t=int(subword_hextobin, 2)^int(rcon_hextobin,2) #xor 
  t=hex(t)[2:]
  #t=groupbytes(t)
  return t
```

```
def generate_40subkey(word_matrix,rcon):
  #temp=[]
  t_values=['-']
  for i in range(4,44):
    if i%4 == 0:
      rw=rotword(word_matrix,i)
      #print('rotword',i,rw)
      sub_list=subword(rw)
      #print('sublist',i,sub_list)
      t=xor(sub_list,rcon[int(i/4)]) #t in hex
      if len(t)<8:
        t='0'+t
      t_values.append(t)

      #print('t',i,t)
      w_hex_string=xor(''.join(word_matrix[i-4,:]),t)
      w_byte=groupbytes(w_hex_string)
      #print('word',i,w_hex_string)
      #print([w_byte])
      word_matrix=np.append(word_matrix,[w_byte], axis=0)
      
    else:
       w_hex_string=xor(''.join(word_matrix[i-4,:]),''.join(word_matrix[i-1,:]))
       if len(w_hex_string)<8:
         w_hex_string='0'+w_hex_string
       w_byte=groupbytes(w_hex_string)
       #print('word',i,w_hex_string)
       #print([w_byte])
       word_matrix=np.append(word_matrix,[w_byte], axis=0)
       #np.vstack ((word_matrix,np.array(w_byte)) )
      
    
    
  return word_matrix,t_values
```
