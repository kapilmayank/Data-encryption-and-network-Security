```
import random
```

```
def phi(p,q):
  return (p-1)*(q-1)
```

```
#function computes gcd(i.e r1) and multiplicative inverse(i.e t1) using extended euclidean algorithm
def emi(r1,r2,s1=1,s2=0,t1=0,t2=1):
    if r2==0:
        return [r1,s1,s2,t1,t2]
    else:
        return emi(r2,r1%r2,s2,s1-int(r1/r2)*s2,t2,t1-int(r1/r2)*t2)
```

```
#public key e
def generate_e(phi_v):
  key=[]
  for i in range(phi_v):
    gcd=emi(i,phi_v,s1=1,s2=0,t1=0,t2=1)
    if gcd[0] is 1:
      key.append(i)
      #print(random.choice(number_list))
  return key,random.choice(key)
```

```
#function to  bring mI in mod range
def mode_range(mod_value,result):
  if 0<result<mod_value:
    return result
  elif result<0:
    results=mod_value+result
    return results
```

```
#private key d
def generate_d(r,phi_v):
  d=emi(phi_v,r,s1=1,s2=0,t1=0,t2=1)#e%phi_v
  d=mode_range(phi_v,d[3])
  return d
```

```
#encryption is done using private key (so this encrypted text is known as digital signature)
def encrypt(m,prkey,n):
  return (m**prkey)%n
```

```
def decrypt(c,pukey,n):
  return (c**pukey)%n
```

```
def read_file(m):
  l_ascii=[]
  orig_message=''
  f = open(m+".txt", "r")
  for i in f.read():
    #print(ord(i))
    orig_message+=i
    l_ascii.append(ord(i))
  return l_ascii,orig_message
 
```

```
l,orig_message=read_file('rsa')
print(l)
print(orig_message)
```

```
#choose 2 large primes
p=7
q=11
n=p*q
phi_v=phi(p,q)
print('Euler totient',phi_v)
k,r=generate_e(phi_v)
print('Possible values of public key',k)
print('Randomly chosen public key',r)#randomly chosen value of public key(e) from the possible values of e 
#print(key)
print('Public key',r)
d=generate_d(r,phi_v)
print('Private key',d)
l,orig_message=read_file('rsa')#returns list of ascii values and original message
cipher=[]
c_ascii=[]
dec=[]
for i in l:
  if i>n:
    i=i-n
    c_asc=encrypt(i,r,n)
    c_ascii.append(encrypt(i,r,n))
    cipher=[chr(i) for i in c_ascii]
    cipher=''.join(cipher)
    de=decrypt(c_asc,d,n)
    de=chr(de+n)
    dec.append(de)
    x=''.join(dec)
  else:
    c_asc=encrypt(i,r,n)
    c_ascii.append(encrypt(i,r,n))
    cipher=[chr(i) for i in c_ascii]
    cipher=''.join(cipher)
    de=decrypt(c_asc,d,n)
    dec.append(chr(de))
    x=''.join(dec)
print(' Alice creates her digital signature using her private key\n',cipher)
print('Alice sends message and signature to Bob........')
print('Original Message \n',orig_message)
print('Bob decrypts the signature using public key of Alice',x)
#x is decrypted message using public key
if orig_message==x:
  print('Bob  accepts the data sent by Alice')
else:
  print('Bob suspects some attack')
```
