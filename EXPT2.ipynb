```
import string
all_letters= string.ascii_letters 
print(all_letters)
print(len(all_letters))
```

```
class Substitution_Cypher:
  #Constructor for intializing the encryption and decryption key which is same here
  def __init__(self,key=3):
    self.key=key

  #This instance  function will read the file line by line and return it in string format 
  def read_plain_txt(self,path):
    plain_txt=''
    file_en = open(path,"r") 
    for char in file_en.readlines():
      plain_txt=plain_txt+char
      return plain_txt
  
  """This will encrypt the file when the user will pass the key
  unlike the original approach where first the corresponding cyphers of each alphabet was stored in the 
  dictionary  and then asigned to plain text in diff loop.Now it is done in a single step
  by extracting the index of every character of plaintext in (all letters) using .index()
  and directly replacing that index by  the new index """
  def encrypt(self,plain_txt,key=3):
    cipher = ''
    for i in plain_txt:
      if i in all_letters:
        cipher+=all_letters[(all_letters.index(i)+key)%len(all_letters)]
      else:
        cipher+=i
    return cipher

  def decrypt(self,cipher,key=3):
    decipher = ''
    for i in cipher:
      if i in all_letters:
        decipher+=all_letters[(all_letters.index(i)-key)%len(all_letters)]
      else:
        decipher+=i
    return decipher
  
  def brute_force_attack(self,cipher):
    from enchant.checker import SpellChecker
   
    for key in range(len(all_letters)):
      decipher=self.decrypt(cipher,key)
      #print(decipher)
      #cracked=decipher.split(' ')
      #out = map(lambda x:x.lower(),cracked)
      #out=list(out)
      d = SpellChecker("en_US")
      d.set_text(decipher)
      errors = [err.word for err in d]
      if len(errors) is 0:
        print('Key found after attack',key,all_letters[key])
        break
```

```
class Transposition_Cypher:
  def encrypt(self,plain_txt,key):
    cipher=''
    for i in range(len(plain_txt)):
      cipher += plain_txt[(i-key)%len(plain_txt)]
    return cipher
    

  def decrypt(self,plain_txt,cipher,key):
    decipher=''
    for i in range(len(plain_txt)):
      decipher += cipher[(i+key)%len(plain_txt)]
    return decipher
  
  def brute_force_attack(self,cipher):
    from enchant.checker import SpellChecker
   
    for key in range(len(all_letters)):
      decipher=self.decrypt(plain_txt,cipher,key)
      d = SpellChecker("en_US")
      d.set_text(decipher)
      errors = [err.word for err in d]
      if len(errors) is 0:
        print('Key found after attack',key,all_letters[key])
        break
 
```
