```
import random
import numpy as np
```

```
# Function to convert decimal to 8 bit binary 
def bin(n):
    b=''
    while True:
        q=int(n/2)
        r=n%2
        b+=str(r)
        n=q
        if n is 0:
            break
    if len(b)<8:
            zeros='0'*(8-len(b))
            return zeros+b[::-1]
    return b[::-1]
```

```
# Function to generate 64 bit random key
def random_key_generator():
  key64=''
  for i in range(8):
    r = random.randint(0, 255)
    b=bin(r)
    key64+=b
  return key64
```
