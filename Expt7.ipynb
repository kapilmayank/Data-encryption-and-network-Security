```
# function to find extended multiplicative inverse
def emi(r1,r2,s1=1,s2=0,t1=0,t2=1):
    if r2==0:
        return [r1,s1,s2,t1,t2]
    else:
        return emi(r2,r1%r2,s2,s1-int(r1/r2)*s2,t2,t1-int(r1/r2)*t2)
   
    
```

```
#function to  bring mI in mod range
def mode_range(mod_value,result):
  if 0<result<mod_value:
    return result
  elif result<0:
    results=mod_value+result
    return results

```

```
#function to find crt(chinese remainder theorem)
def crt(a1,m1,a2,m2,a3,m3):
  M=m1*m2*m3
  print('M =',M)
  M1=int(M/m1)
  M2=int(M/m2)
  M3=int(M/m3)
  i1=emi(m1,M1,s1=1,s2=0)
  inv1=mode_range(m1,i1[3])
  print('M1 inverse',inv1)
  i2=emi(m2,M2,s1=1,s2=0)
  inv2=mode_range(m2,i2[3])
  print('M2 inverse',inv2)
  i3=emi(m3,M3,s1=1,s2=0)
  inv3=mode_range(m3,i3[3])
  print('M3 inverse',inv3)
  x=(a1*inv1*M1+a2*inv2*M2+a3*inv3*M3)%M
  x=mode_range(M,x)
  print('x=',x)  


```

```
crt(3,7,3,13,0,12)
```

```
crt(2,3,3,5,2,7)
```

Conclusion:
Chinese remainder theorem helps in finding the solution(i.e x) efficiently when a set of equations have the same result after performing mod operation.It is used in various cyphers in cryptography
